```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./data/train.csv')

df_nan_bad = df.loc[df.bad.isna()]

df_nan_bad.reset_index(inplace=True)

impotant_props = ['count_overdue_all_3lm', 'work_code', 'all_creds_count_lm',
       'mfo_inqs_count_month', 'bank_inqs_count_quarter',
       'mfo_closed_count_ly', 'all_creds_count_all', 'delay_more_sum_all',
       'region', 'cred_day_overdue_all_sum_all']
```

```python
sns.distplot(df_nan_bad['count_overdue_all_3lm'], label='nans')
sns.distplot(df['count_overdue_all_3lm'], label='all')
plt.title('count_overdue_all_3lm')
plt.legend()
plt.show()
```
